## Passo 1: Importando o pandas

O **pandas** é o principal pacote Python para manipulação de dados tabulares. A convenção é importá-lo com o alias `pd`.

```python
import pandas as pd
```

## Passo 2: Carregando o arquivo de dados

O arquivo `default_of_credit_card_clients__courseware_version_1_21_19.xls` contém os dados do estudo de caso. Usamos `pd.read_excel()` para carregá-lo.

> O parâmetro `header=1` indica que a linha de cabeçalho real está na segunda linha da planilha.

```python
df = pd.read_excel(
    'default_of_credit_card_clients__courseware_version_1_21_19.xls',
    header=1
)
```

## Passo 3: Examinando os dados carregados

Verificamos o número de linhas e colunas e examinamos os primeiros registros para confirmar que o carregamento foi feito corretamente.

```python
# Número de linhas e colunas
print(f"Linhas: {df.shape[0]}")
print(f"Colunas: {df.shape[1]}")
```
